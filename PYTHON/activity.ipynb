```python
lower_limit = df['actions_per_user'].quantile(0.10)
upper_limit = df['actions_per_user'].quantile(0.90)
```

```python
print(lower_limit, upper_limit)
```

```python
def activity(a):
    if a <= lower_limit:
        return 'low_activity'
    elif a >= upper_limit:
        return 'high_activity'
    else:
        return 'normal'
```

```python
df['activity_status'] = df['actions_per_user'].apply(activity)
```

```python
df['activity_status'].value_counts()
```
